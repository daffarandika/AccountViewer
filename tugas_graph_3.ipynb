```python
G = nx.Graph()
G.add_edge(1,2,weight=10)
G.add_edge(1,5,weight=45)
G.add_edge(1,4,weight=30)
G.add_edge(2,3,weight=50)
G.add_edge(2,5,weight=40)
G.add_edge(2,6,weight=25)
G.add_edge(3,5,weight=35)
G.add_edge(3,6,weight=15)
G.add_edge(4,6,weight=20)
G.add_edge(5,6,weight=55)

labels=nx.get_edge_attributes(G, 'weight')

pos={
    1:(0,5.8),
    2:(5.8,5.8),
    3:(9,4.5),
    4:(0,3.5),
    5:(6,3),
    6:(2,0.1)
}
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
```
